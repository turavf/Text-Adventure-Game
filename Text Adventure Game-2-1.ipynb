```python

"""
    DocString:
    
    A) Introduction:
    This game is based on the movie Angels & Demons(1975), the sequel to the
    2006 film The Da Vinci Code (in spite of the movie being a sequel,
    the book was meant to be a prequel, hence it contains no spoilers on the
    The Da Vinci Code movie). 
    It consists of four stages, and also has defined functions
    for start, win, and lose. This game requires honesty and an intention to
    have a great time (having seen the movie will be helpful).
    *The game is not 100% accurate to the movie for the purpose of not being 
    too long.
    
    B) Known Bugs and/or Errors:
    None.
    
"""

#Importing packages for later use
from sys import exit
import random

#Stage 1
#Through the course of this code, we define a lot of variables to put more emphasis on the different scenarios inside one stage
#setting the context of the game for the user
def context():
    print("""
    The Ring of the Fisherman, which bears the official papal seal,
    must be destroyed immediately following the pope's death.
    The papal apartment is then sealed for nine days of mourning,
    a period known as sede vacante: The time of the empty throne.
    """)
    input(prompt="<Press enter to continue>\n")
    introduction()

#defining the introduction and the basis for the game
def introduction():
    print("""
    Four cardinals were kidnapped just hours after the death of the pope.
    Shortly afterwards you recieve an envelope from the Vatican.
    
    You have two options:
    1) Open the envelope.
    2) Leave the cardinals to die.
    """)

#user input to give the user the chance to decide if they want to play the game or exit
    choice = input(prompt=">  \n")
#Using case sensitivity control casefold for the input
    choice.casefold()

#Using conditional statements for simple choices
    if "1" in choice or "open" in choice or "envelope" in choice or "one" in choice:
        print("\nYou open the envelope\n")
        input(prompt="<Press enter to continue>\n")
        beginning()

    #Exit option in case the user wants to leave the game
    elif "2" in choice or "Leave" in choice or "cardinals" in choice or "die" in choice or "two" in choice:
        print("\nYou have decided not to pursue the quest. You can continue your day.")
        exit(0)

    #else clause in case the user prints a command outside the options given
    else:
        print("Invalid command. Please try again.")
        input(prompt="<Press enter to continue>\n")
        introduction()

#second part of the introduction
def beginning():
    print("""
    The envelope contains a paper with a word: Illuminati, and a note that 
    states the four favorite cardinals to be elected pope, 
    will be publicly executed, one per hour, beginning today.
    
    You have two options:
    1) Go to the Vatican.
    2) Leave the cardinals to die.
    """)
    
    #giving the user another chance to decide if they want to play the game or exit
    choice = input(prompt=">  \n")
    #Using case sensitivity control casefold for the input
    choice.casefold()

#Using conditional statements for simple choices
    if "1" in choice or "go" in choice or "Vatican" in choice or "one" in choice:
        print("\nYou are on a plane to the Vatican\n")
        input(prompt="<Press enter to continue>\n")
        Vatican()
    
    #option that directs the user to exit
    elif "2" in choice or "Leave" in choice or "cardinals" in choice or "die" in choice or "two" in choice:
        print("\nYou have decided not to pursue the quest. You can continue your day.")
        exit(0)

    #else clause in case the user prints a command outside the options given
    else:
        print("Invalid command. Please try again.")
        input(prompt="<Press enter to continue>\n")
        beginning()
        
#Stage 2
#user defined functions to state the main argument of the game
def Vatican():
    print("""
    You arrive at the Vatican where you are greeted by the Inspector general 
    of the Vatican police force.
    He tells you they received a video from the kidnapers.
    """)
    input(prompt="<Press enter to listen video>\n")
    First_Enigma()

def First_Enigma():
    print("""
    We will destroy your four pillars, 
    we will brand your preferiti and 
    publicly sacrifice them on the altars of science. 
    A shining star at the end of the Path of Illumination.
    """)
    input(prompt="<Press enter to continue>\n")
    print("""
        After examining the enigma, you deduce the cardinals will be killed 
        at the four altars of the Path of Illumination, each one represented
        by one of the four elements: Earth, Air, Fire and Water.
        """)
    input(prompt="<Press enter to continue>\n")
    Archives()
    
#Stage 3
def Archives():
    print("""
        Since you are at the Vatican, 
        you decide to go to their archives to find some information.
        However they forgot to give you the password to enter the archives.
        """)
    input(prompt="<Press enter to discover the password>\n")
    bingo_game()

#Creating a short game of Bingo
#*** The game of Bingo is said to have been originated in Italy during the XVI century,
# that was the same time as the Illuminati and the artist in which this movie is based.
def bingo_game():
    random_draw_list = random.sample(range(1, 21), 20)

    #Generating and printing a card for the user with the numbers to play
    def bingo_card():
        card = {
            "P": [],
            "A": [],
            "S": [],
        }
        min = 1
        max = 10
        #using a loop and the random package to make the game different every time the user plays
        for letter in card:
            card[letter] = random.sample(range(min, max+1),3)
            min += 1
            max += 1
            if letter == "N":
                card[letter][2] = "X"
        return card
    
    #printing the card
    def print_card(card):
        for letter in card:
            print(letter, end="\t")
            for number in card[letter]:
                print(number, end="\t")
            print("\n")
        print("\n")

    #Drawing a random number
    def draw(card, list):
        number_drawn = random_draw_list.pop()
        for letter in card:
            i = 0
            for number in card[letter]:
                if number == number_drawn:
                    card[letter][i] = "X"
                i += 1
        return number_drawn

    #Defining winning rules according to Bingo:
    def winning(card):
        win = False
        #checking the numbers from the diagonal
        if card["P"][0] == "X" and card["A"][1] == "X" and card["S"][2] == "X":
            win = True
        #checking the numbers from the other diagonal
        elif card["S"][0] == "X" and card["A"][1] == "X" and card["P"][2] == "X":
            win = True
        #checking the corners
        elif card["P"][0] == "X" and card["S"][2] == "X" and card["P"][2] == "X" and card["S"][0] == "X":
            win = True
        #checking the horizontal and vertical axis
        for letter in card:
            if(len(set(card[letter]))==1):
                win = True
        for i in range(3):
            cnt = 0
            for letter in card:
                if card[letter][i] == "X":
                    cnt += 1
            print(cnt)
            if cnt == 3:
                win = True
                #breaking the loop when the user wins
                break
        return win

    #Defining the game
    def function():
        print("Let's play bingo!")
        card = bingo_card()

        print("\nHere is your card:\n")
        print_card(card)

        print("\nKeep pressing enter to continue or type quit to exit.\n")
        user = input(prompt=">  \n")
        win = winning(card)
        balls_drawn = 0

        #while loop to play the game until winning
        while win == False and user != "quit":
            number_drawn = draw(card, random_draw_list)
            #counting the number of balls drawn
            balls_drawn += 1

            #printing the number drawn
            print(f"\nNumber drawn: {number_drawn}.")
            print(f"Total balls drawn: {balls_drawn}.\n")
            print_card(card)

            win = winning(card)

            user = input(prompt=">  \n")
       
        #printing winning and losing sentence
        if win:
            print(f"\nBingo! You have discovered the password!")
        else:
            print("You have lost the game, please try again")
    function()

    Second_Enigma()
    
#Stage 4
def Second_Enigma():
    #printing the information of the second enigma
    print("""
    You've managed to enter the Archives.
    
    You stumble a volume written by Galileo, a known member of the Illuminati.
    Inside you find an enigma written by the Illuminati.
    """)
    input(prompt="<Press enter to read the enigma>\n")

    print("""
    From Santi's Earthly tomb with Demons Hole,
    Across Rome the mystic elements unfold,
    Woman left completely afire,
    The angel of peace will perspire.
    The path to the center is laid, the sacred test,
    Let angels guide you on your lofty quest.
    """)
    input(prompt="<Press enter to continue>\n")
    print("""
    The enigma contains clues to find the four altars. 
    The first one is Earth, followed by Air, Fire and finally Water.
                    """)
    input(prompt="<Press enter to solve it.>\n")
    Earth()

#user defined function for the first minigame
def Earth():
    print("""
        Earth riddle.
                    """)
    input(prompt="<Press enter to continue>\n")
    print("""
        You remember Santi is the last name of Raphael, a famous Italian sculptor.
        Raphael is buried at the pantheon, the oldest Catholic church in Rome.
                    """)
    input(prompt="<Press enter to continue>\n")
    print("""
         1) Go to the Pantheon.
         2) Think more about the riddle.
                    """)
    
    #input for the user to choose the answer to the riddle
    choice_earth = input(prompt=">  \n")
    choice_earth.casefold()

    #Since we are at the center of the game, we put the function fail when the user makes a bad choice
    if "1" in choice_earth or "go" in choice_earth or "Pantheon" in choice_earth or "one" in choice_earth:
        print("""
            You arrive at the Pantheon. 
            Unfortunately, you got the riddle wrong. 
            You spend too much time here.
                          """)
        fail()
    
    #correct choice in the elif clause
    elif "2" in choice_earth or "Think" in choice_earth or "more" in choice_earth or "riddle" in choice_earth or "two" in choice_earth:
        print("""
            You realize maybe the riddle is talking about a chapel he built,
            not the one he was buried in.
            You also realize that demon's hole could be a reference to 
            a crypt.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            After looking it up on the internet, you discover he only designed
            one undercroft: The Chigi Chapel; which was once called, 
            The Capella della Terra (The Chapel of the Earth).
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to go to the chapel.>\n")
        print("""
            You rush to the chapel. But it's empty...
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue.>\n")
        print("""
            You see a mosaic in the floor with a drawing of the devil
            and the words: Mors ad caelos.
                    
            You found the demon's hole!
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue.>\n")
        print("""
            You break it and inside you find a secret passage.
            At the end of the passage you see a figure.
                    
            You've found the cardinal!
                          """)
        input(prompt="<Press enter to continue.>\n")
        Air()

    #else clause in case the user prints a command outside the options given
    else:
        print("Invalid command. Please try again.")
        input(prompt="<Press enter to try again>\n")
        Earth()

#user defined function for the second minigame
def Air():      
    print("""
        Air riddle.
                    """)
    input(prompt="<Press enter to continue>\n")
    print("""
        You realize the verse:
        The path to the center is laid.
                    
        Could be referring to the central plaza of the Vatican:
        Saint Peter's Square.
                    """)
    #using input so the user keeps interacting with the game in between long text
    input(prompt="<Press enter to continue>\n")
    print("""
         You arrive at the Square, however, it's full of people
         waiting for the revealing of the new pope and you can barely
         see anything.
                    """)
    #using input so the user keeps interacting with the game in between long text
    input(prompt="<Press enter to continue>\n")
    print("""
       There's only 8 minutes left until the next execution, 
       what do you do?
                    
        1) Sit on the floor
        2) Try to get through the people
        3) Abandon this location and look for another plaza
                    """)
    
    #input for the user to write his answer to the riddle
    choice_air = input(prompt=">  \n")
    #Using case sensitivity control casefold for the input
    choice_air.casefold()

    #conditional statements to define different stages depending on the user's answer
    if "1" in choice_air or "Sit" in choice_air or "floor" in choice_air or "one" in choice_air:
        print("""
            You decide to rest on the floor.
                    
            Suddenly, you see a Bas-relief with a drawing of a man blowing.
                     """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            You notice the man is looking southwest so you follow that direction.
            You encounter another Bas-relief and again you follow the direction.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            You arrive at the sculpture of an angel. The cardinal is laying underneath,
            tied with a bag asphyxiating him.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            It looks like you've arrived on time! The cardinal is recovering.
                          """)
        input(prompt="<Press enter to continue>\n")
        Fire()

    #elif choice to give an alternative storyline that also moves to the next stage
    elif "2" in choice_air or "Try" in choice_air or "through" in choice_air or "people" in choice_air or "two" in choice_air:
        print("""
            There's two many people.
                    
            You can't get through.
                    
            Time is running out.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            You decide to rest on the floor.
                    
            Suddenly, you see a Bas-relief with a drawing of a man blowing.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            You notice the man is looking southwest so you follow that direction.
            You encounter another Bas-relief and again you follow the direction.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            You arrive at the sculpture of an angel. The cardinal is laying underneath,
            tied with a bag asphyxiating him.
                          """)
        #using input so the user keeps interacting with the game in between long text
        input(prompt="<Press enter to continue>\n")
        print("""
            It looks like you've arrived on time! The cardinal is recovering.
                          """)
        input(prompt="<Press enter to continue>\n")
        Fire()
    
    #elif option that leads to failing and thus ending the game
    elif "3" in choice_air or "Abandon" in choice_air or "location" in choice_air or "look" in choice_air or "another" in choice_air or "plaza" in choice_air or "three" in choice_air:
        print("""
            You were in the right place. 
            Your time is up and you can't save the cardinal.
                          """)
        fail()

    #else clause in case the user prints a command outside the options given
    else:
        print("Invalid command. Please try again.")
        input(prompt="<Press enter to try again>\n")
        Air()

#user defined function for the third minigame
def Fire():             
    print("""
        Fire riddle.
                    """)
    input(prompt="<Press enter to continue>\n")
    print("""
        You assume the element fire is related to the verse:
        Woman left completely afire.
                    """)
    input(prompt="<Press enter to continue>\n")
    print("""
        After thinking carefully, you arrive to the conclusion
        that it can only be the sculptures of two woman:
                    
        1) Seraphim the fiery one.
        2) The ecstasy of Saint Teresa.
                    """)
    
    #input for the user to write his choice to the riddle
    choice_fire = input(prompt=">  \n")
    #Using case sensitivity control casefold for the input
    choice_fire.casefold()

    #scenario for the first choice that leads to failing the game
    if "1" in choice_fire or "Seraphim" in choice_fire or "fiery" in choice_fire or "one" in choice_fire:
        print("""
            You chose wrong. 
            Your time is up and you can't save the cardinal.
                          """)
        fail()
    
    #another scenario resulting from the second option
    elif "2" in choice_fire or "ecstasy" in choice_fire or "Saint" in choice_fire or "Teresa" in choice_fire or "two" in choice_fire:
        print("""
            You are on your way to Santa Maria della Vittoria, were 
            the statue lays.
                          """)
        input(prompt="<Press enter to continue>\n")
        #stating list of options
        print("""
            You suddenly realize fire is a dangerous element and perhaps
            you need help.
                    
            You are indecisive about what to bring. 
            Here are your options:
            1) Water
            2) Knife
            3) Firefighters
            4) Bandages
            5) Ointment
                          """)

        #controlling for errors related to the list with the function try and except       
        try:
            choice_ = input(prompt="<Enter an item from the list>\n")
            #Using case sensitivity control casefold for the input
            choice_.casefold()

            #creating a list to store the options
            lst = ['Water', 'Knife', 'Firefighters', 'Bandages', 'Ointment']

            #for loop to see the different outcomes regarding each object of the list 
            for item in lst:
                #choice that leads to moving forward with the game
                if choice_ == 'Knife' or choice_ == '2' or choice_ == 'two':
                    print("""
                        You arrive at the church and find the cardinal
                        hanging over a huge bonfire.
                        You chose wisely and you can cut the ropes with
                        the knife.
                        You saved the cardinal!
                                """)
                    Water()
                #choice that leads to moving forward with the game
                elif choice_ == 'Firefighters' or choice_ == '3' or choice_ == 'three':
                    print("""
                        You arrive at the church and find the cardinal
                        hanging over a huge bonfire.
                        You chose wisely and the firefighters manage to
                        put out the fire.
                        You saved the cardinal!
                                """)
                    Water()
                #wrong choice that leads to failing the game
                elif choice_ == 'Water' or choice_ == '1' or choice_ == 'one':
                    print("""
                        You arrive at the church and find the cardinal
                        hanging over a huge bonfire.
                        The fire is too big and you can't put it out.
                        The cardinal dies!
                                """)
                    fail()
                #wrong choice that leads to failing the game
                elif choice_ == 'Bandages' or choice_ == 'Ointment' or choice_ == '4' or choice_ == '5' or choice_ == 'four' or choice_ == 'five':
                    print("""
                        You arrive at the church and find the cardinal
                        hanging over a huge bonfire.
                        You can't put it out with the item you chose
                        The cardinal dies!
                                """)
                    fail()
                    
                #else clause in case the user prints a command outside the options given
                else:
                    print("Invalid command. Please try again.")
                    input(prompt="<Press enter to continue>\n")
                    Fire()
                    
        #ending the control for errors related to the list with the function try and except            
        except:
            print("You need to input an item from the list")

    #else clause in case the user prints a command outside the options given
    else:
        print("Invalid command. Please try again.")
        input(prompt="<Press enter to try again>\n")
        Fire()

#user defined function for the last minigame
def Water():  
    print("""
        Water riddle.
                    """)
    input(prompt="<Press enter to continue>\n")
    print("""
        You assume the element water is related to the verse:
        The angel of peace will perspire.
                    
        Any guesses on what the angel of peace could be?
                    """)
    
    #input for the user to write his answer
    answer = input(prompt="<Press enter your answer>\n")
    #Using case sensitivity control casefold for the input
    answer.casefold()
    
    #using an if/else clause for the correct and incorrect answer
    if answer == "dove" or answer == "Dove" or answer == "the" or answer == "a":
        print("You're right!")
                    
    else:
        print("""
            Your answer was wrong, the solution was The Dove.
                    """)    
    input(prompt="<Press enter to continue>\n")
    print("""
        Since the element is water, you look for a fountain with a Dove.
        There is only one in Rome: The fountain of the 4 rivers.
                    """)
    #using input so the user keeps interacting with the game in between long text
    input(prompt="<Press enter to continue>\n")
    print("""
        You rush to the fountain and find the cardinal in the bottom
        of the fountain tied to a chair.
        You arrived on time and you saved him!
                    """)
    ending()

#Printing and ending statement to wrap up the game
def ending():
    print("""
    You have found the four cardinals! 
    Now the Conclave can begin and we'll soon have a new pope!
          """)
    input(prompt="<Press enter to continue>\n")
    win()
    
#Defining winning function
def win():
    print("""
    You have completed your quest and saved the cardinals, Great job!🎉
          """)
    input(prompt="<Press enter to exit>\n")
    exit(0)
    
#Defining failing function
def fail():
    print("You have failed in your quest, the cardinals are dead😔.")
    input(prompt='<Game Over>')
    exit(0)

#Starting the game
context()
```
